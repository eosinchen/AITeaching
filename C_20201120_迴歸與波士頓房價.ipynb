```
from sklearn import datasets  # 引入sklearn裏頭的資料集
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd # 引入Pandas

%matplotlib inline
# %matplotlib inline是一個魔法函式。官方給出的定義是：IPython有一組預先定義好的所謂的魔法函式，可以通過命令列的語法形式來訪問它們。
# %matplotlib inline 可以在Ipython編譯器比如jupyter notebook 或者 jupyter qtconsole裡直接使用，功能是可以內嵌繪圖，並且省略掉plt.show()。

data = datasets.load_boston() # 取得波士頓房價的數據
data # 查看原始資料,以 JSON 方式存在
```

觀察每個keys：

DESCR：這個資料集的描述

data：每個房子的資訊

feature_names：每個房子的特徵

target：每個房子的價格

filename: 檔案名稱

```
print(data.keys())
```

每個欄位的詳細描述如下：

```
print(data.DESCR)
```

把資料轉換成 pd.DataFrame

```
df_data = pd.DataFrame(data.data, columns=data.feature_names) # 將數據以改成DataFrame的方式呈現
df_data #　查看格式化後的資料
```

```
df_data['MEDV'] = data.target
df_data
```

# **資料探索**

資料視覺化

```
# 用 seaborn 一次把圖表的視覺化格式設定好，這裡是只有先設定圖表長寬
sns.set(rc={'figure.figsize':(10,10)})

# 使用的資料是房價MEDIV
sns.distplot(df_data['MEDV'])

plt.show()
```

下一段程式碼：[關聯陣列]

每個變數之間的關係，透過相關係數去觀察有哪些特徵變數和目標變數有較高的相關性等等：

跟MEDV（房價）高度相關的是LSTAT（中低收入戶佔當地居住人口的比例），和RM（房子有幾間房間）這兩個變數。

此外也看到DIS（到波士頓商業中心的距離）和AGE（屋齡），INDUS（非零售業土地使用比例）和ZN（居住使用土地比例）這兩組變數有多元共線性問題，所以未來如果要做其他模型，避免同時使用這兩組中的變數。

```
correlation_matrix = df_data.corr().round(2)

# annot = True 讓我們可以把數字標進每個格子裡
sns.heatmap(data=correlation_matrix, annot = True)
```

目前可以用LSTAT和RM來做出預測MEDV的模型。再次把這兩個變數跟房價變數的關係畫出來，可以看到兩者和房價變數都接近線性關係：

```
# 設定整張圖的長寬
plt.figure(figsize=(20, 5))
features = ['LSTAT','RM']
target = df_data['MEDV']
for i, col in enumerate(features):
 # 排版1 row, 2 columns, nth plot：在jupyter notebook上兩張並排 
 plt.subplot(1, len(features) , i+1)
 # add data column into plot
 x = df_data[col]
 y = target
 plt.scatter(x, y, marker='o')
 plt.title(col)
 plt.xlabel(col)
 plt.ylabel('MEDV')
```
